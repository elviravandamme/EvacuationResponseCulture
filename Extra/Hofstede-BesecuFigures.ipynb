```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_hofstede = pd.read_excel('Hofstede_scores_4landen.xlsx')
```

```python
sns.set(font_scale=2 )
sns.set_palette("tab10")
sns.set_style("white")
#sns.set_palette("bright")
basicplot= sns.catplot(x="Cultural dimension", y="Score", hue="Culture", kind="bar", data=df_hofstede, height=8, aspect= 15/7,  hue_order=["Czech Republic", "Poland", "Turkey", "UK"])
plt.xticks(rotation=45, horizontalalignment='right') #, fontsize='xx-large'
basicplot.fig.subplots_adjust(top=.9)
basicplot.fig.suptitle('Cultural dimension scores for Czech Republic, Poland, Turkey and the UK')
ax = basicplot.facet_axis(0,0)
basicplot.set(xlabel=None)

basicplot.set(ylim=(0, 100))
for p in ax.patches:
    ax.text(p.get_x() , 
            p.get_height() , 
            '{0:0.0f}'.format(p.get_height()), 
             rotation='horizontal', size='small', )
    
#plt.savefig("Dimensionscores.png", dpi=300)
```

```python
df_besecu = pd.read_excel('Data Besecu.xlsx')
```

```python
sns.set(font_scale=2 )
sns.set_palette("tab10")
sns.set_style("white")
#sns.set_palette("bright")
basicplot= sns.catplot(x="Task", y="Number", hue="Culture", kind="bar", data=df_besecu, height=8, aspect= 15/7,  hue_order=["Czech Republic", "Poland", "Turkey", "UK"])
#plt.xticks(rotation=45, horizontalalignment='right') #, fontsize='xx-large'
basicplot.fig.subplots_adjust(top=.9)
basicplot.fig.suptitle('Number of tasks performed during the BeSeCu experiments')
ax = basicplot.facet_axis(0,0)

basicplot.set(xlabel=None)
basicplot.set(ylabel="Number of tasks")



for p in ax.patches:
    ax.text(p.get_x() , 
            p.get_height() , 
            '{0:0.01f}'.format(p.get_height()), 
              size='small',horizontalalignment= "left", verticalalignment="bottom")
    
```
